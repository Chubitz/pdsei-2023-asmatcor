* **Librerias a utilizar:**

```
import numpy as np
import matplotlib.pyplot as plt
import pylab as pl
import time
from IPython import display
plt.style.use(['dark_background']) #Para graficas para temas oscuros
```

###**4.2. Cálculo de la convolución usando una funcion de alto nivel**

![https://graphics.stanford.edu/courses/cs178/applets/convolution-equation2-c.jpg](https://graphics.stanford.edu/courses/cs178/applets/convolution-equation2-c.jpg)


![https://qph.fs.quoracdn.net/main-qimg-cb81d3dfbe146b84ebf717ec12722f32.webp](https://qph.fs.quoracdn.net/main-qimg-cb81d3dfbe146b84ebf717ec12722f32.webp)

* **Creación de la señal y el kernel**

```
#Creando una señal arbitraria

senal1 = np.concatenate((np.zeros(30),np.ones(2),np.zeros(20),np.ones(30),2*np.ones(10),np.zeros(30),-np.ones(10),np.zeros(40)),axis=0)

kernel = np.exp(-np.linspace(-2,2,20)**2) #Funcion gaussiana
kernel = kernel/sum(kernel)

N = len(senal1) #Longitud de la señal
```

* **Mostrando señales creadas**

```
plt.figure(figsize=(15,6)) #Estableciendo tamaño de la grafica

#Grafice del Kernel

plt.subplot(211)
plt.plot(kernel,'r', linewidth=2)
plt.xlim([0,N]) #Limitamos el eje x entre 0 y N
plt.title('kernel')
plt.grid

#Gréfica de la señial
plt.subplot(212)
plt.plot(senal1,'b',linewidth=2,)
plt.xlim([0,N]) #Limitamos el eje x entre 0 y N
plt.title('señal 1')
plt.grid()

plt.show()
```

 * **Cálculo y gráfica de la convolución**

```
resultado = np.convolve(senal1, kernel, 'same')

plt.figure(figsize=(15,3))

plt.plot(resultado,'y',linewidth=2)
plt.xlim([0, N])
plt.title('Resultado de la convolución')
plt.grid()

plt.show()
```

###**4.2. Detalles de la convolución**
>*Longitud de la convolución de 2 señales.*

>*Efecto del kernel*

* **Creación de señales y aplicación de la convolución**





```
#Creando un pulso
senal1 = np.zeros(20)
senal1[8:15] = 1

# Kernel arbitrario
kernel = np.array([1,.8,.6,.4,.2])

# Tamaño de la convolucion
nSenal1 = len(senal1)
nKernel = len(kernel)
nConvolucion = nSenal1 + nKernel - 1

#Aplicacion de la convolucion
resultado = np.convolve(senal1, kernel, 'full')
```

* **Graficando resultados**

```
plt.figure(figsize=(15,9))

#Gráfica de la señal1
plt.subplot(311)
plt.plot(senal1,'ro-',linewidth=2)
plt.xlim([0,nSenal1-1])
plt.title('Señal')
plt.grid()

# Gráfica del kernel
plt.subplot(312)
plt.plot(kernel,'bo-',linewidth=2)
plt.xlim([0,nSenal1-1])
plt.title('Kernel')
plt.grid()

# Gráfica de los resultados
plt.subplot(313)
plt.plot(resultado,'yo-',linewidth=2)
plt.xlim([0,nConvolucion])
plt.title('Resultado de la Convolucion')
plt.grid()
plt.show()
```

###**4.3. Animación de la convolución**

*Se usarán las señales creadas en el apartado anterior*

```
import warnings
warnings.filterwarnings('ignore')
```

```
#Calculamos la mitad del kernel
mitad_kernel = int(np.floor(nKernel/2))

#Invertimos el kernel
kernel_invertido = kernel[::-1]

#Rellenando la señal con 0's para almacenar la salida
senal1_pad = np.concatenate((np.zeros(mitad_kernel), senal1, np.zeros(mitad_kernel)))

#Vector para almacenar el resultado
conv_resultado = np.zeros_like(senal1_pad)

#Correr la convolucion
plt.figure(figsize=(15,5))

for ti in range(mitad_kernel,len(senal1_pad)-mitad_kernel):

  #Extrayendo una ventana del tamaño del kernel
  ventana = senal1_pad[ti-mitad_kernel: ti+mitad_kernel+1]

  #Producto punto de la ventana con el kernel invertido
  conv_resultado[ti] = np.sum(ventana * kernel_invertido)

  #Graficando
  plt.cla() #Limpiar los ejes
  plt.axis(['auto',  'auto', -0.5, 3.2])
  plt.plot(senal1_pad,"go-") #muestra la señal original
  plt.plot(np.arange(ti-mitad_kernel, ti+mitad_kernel+1), kernel_invertido, "ro-") #Muestra el movimiento del kernel)
  plt.plot(np.arange(mitad_kernel,ti), conv_resultado[mitad_kernel:ti], "bo-") #Muestra el resultado
  plt.xlim([0,nConvolucion]) #Rango del eje x

  display.clear_output(wait=True) #Limpia la salida de la celda
  display.display(plt.gcf())

  time.sleep(0.5)

#Recortando los ejes
#conv_resultado = conv_resultado[mitad_kernel:-mitad_kernel]

```

```
plt.figure(figsize=(15,4))
plt.plot(resultado,'b',linewidth=2, label='Funcion convolve()')
plt.plot(conv_resultado,'rs', label='Convolucion manual')
plt.legend()
plt.show()
```

```
!pip install pysoundfile
!pip install bitstring
import IPython
import soundfile
import matplotlib.pyplot as plt
import numpy as np # manejo de arrays
import matplotlib.pyplot as plt # visualizacion de graficos
plt.style.use(['dark_background']) #graficas con fondo oscuro
```

- **Importación de la señal de audio**

```
!gdown --id 1FBGxD-4gnH80V6346RLsAdCuJBaiL0QN

IPython.display.Audio('Cage.wav')
```

- **Creación de la señal y el kernel**

```
audio_stereo, fm = soundfile.read('Cage.wav')
#Obteniendo una señal (mono)
senal = np.mean(audio_stereo, axis=1)

#Creando el kernel
kernel = np.exp(-np.linspace(-2,2,31)**2) #Función gaussiana de 31 muestras
kernel = kernel / sum(kernel)
N = len(senal) #Longitud de la señal
```

- **Mostrando señal creada**

```
plt.figure(figsize=(15,6)) #Tamaño de gráfica

#Gráfica del Kernel
plt.subplot(211)
plt.plot(kernel,'r', linewidth=2)
plt.xlim([0,30]) #Limitando eje X entre (0 - N)
plt.title('Kernel')
plt.grid()

#Gráfica de la señal
plt.subplot(212)
plt.plot(senal,'b', linewidth=2,)
plt.xlim([0,N]) #Limitando eje X entre (0 - N)
plt.title('Señal')
plt.grid()
plt.show
```

- **Calculo y gráfica de la convolución**

```
resultado = np.convolve(senal,kernel,'same') #Convolución
resultado = resultado

plt.figure(figsize=(15,3))

plt.plot(resultado,'y',linewidth=2)
plt.xlim([0,N]) #Limitando eje X entre (0 - N)
plt.title('Resultado de la convolución')
plt.grid()
plt.show()
```

- **Gráfica comparativa entre audio original y audio modificado**


```
plt.figure(figsize=(15,3))
plt.plot(senal, label="Audio Original")
plt.plot(resultado, label="Audio Modificado")
plt.title("Audio Original vs Audio Modificado")
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
plt.xlim([0,N]) #Limitando eje X entre (0 - N)
plt.grid()
plt.legend()
plt.show()
```

- **Resultado de la convolución**

```
soundfile.write('Cage_convolucion.wav', resultado, fm)
IPython.display.Audio('Cage_convolucion.wav')
```

###**2. Test de comprobación**

- **¿Si en el ejercicio 5a usa la función correlate en vez de convolve que diferencias en la forma y sonoras se tendría en la señal de salida?. Explique.**

Las funciones correlate y convolve son utilizadas en procesamiento de señales y procesamiento de imágenes para calcular la correlación y la convolución entre dos secuencias o imágenes respectivamente. La diferencia principal entre ambas radica en cómo se realizan los desplazamientos. En la correlación, se desplaza la segunda señal en el dominio del tiempo, mientras que en la convolución, se invierte antes de desplazarla. Aplicaremos el mismo procedimiento al audio pero utilizando correlación (mismo kernel).

- **Calculo y gráfica de la correlación**

```
resultado_cor = np.correlate(senal,kernel,'same') #Correlación
resultado_cor = resultado_cor

plt.figure(figsize=(15,3))

plt.plot(resultado_cor,'y',linewidth=2)
plt.xlim([0,N]) #Limitando eje X entre (0 - N)
plt.title('Resultado de la correlación')
plt.grid()
plt.show()
```

- **Gráfica comparativa entre audio original y audio modificado**

```
plt.figure(figsize=(15,3))
plt.plot(senal, label="Audio Original")
plt.plot(resultado_cor, label="Audio Modificado con correlación")
plt.title("Audio Original vs Audio Modificado con correlación")
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
plt.xlim([0,N]) #Limitando eje X entre (0 - N)
plt.grid()
plt.legend()
plt.show()
```

- **Resultado de la correlación**

```
soundfile.write('Cage_convolucion.wav', resultado_cor, fm)
IPython.display.Audio('Cage_convolucion.wav')
```

Como se puede observar en la gráfica y escuchar en el audio, que no existe ninguna diferencia entre la señal de salida debido a la correlación y convolución; esto se debe a que como se puede observar en la gráfica del kernel gaussiano esta se trata de una señal simétrica, y al invertirse para realizar la convolución se obtiene la misma señal original.

```
plt.figure(figsize=(15,6)) #Tamaño de gráfica
#Gráfica del Kernel gaussiano
plt.subplot(211)
plt.plot(kernel,'r', linewidth=2)
plt.xlim([0,31]) #Limitando eje X
plt.title('Kernel gaussiano')
plt.grid()


kernel_invertido = kernel[::-1]
#Gráfica del Kernel gaussiano invertido
plt.subplot(212)
plt.plot(kernel_invertido,'b', linewidth=2)
plt.xlim([0,31]) #Limitando eje X entre (0 - N)
plt.title('Kernel gaussiano invertido')
plt.grid()
```

Por lo tanto  no existe ninguna diferencia en la forma y ni en el sonido de la señal de salida producto de la convolución y correlación.
